## **dictionaries**

```python
student ={'name':'John','age':25,'courses':['Math','science']}

print(student)
```

```python
student ={'name':'John','age':25,'courses':['Math','science']}

print(student['courses'])
```

```python
#when key doesnt consist
student ={'name':'John','age':25,'courses':['Math','science']}

print(student['phone'])
```

```python
#get method
student ={'name':'John','age':25,'courses':['Math','science']}

print(student.get('name'))
```

```python
#get method ->when the key is not present
student ={'name':'John','age':25,'courses':['Math','science']}

print(student.get('phone'))
```

```python
#we can also set the default value which values are not exist
#another use of the get method
student ={'name':'John','age':25,'courses':['Math','science']}


print(student.get('phone','Not found'))
```

```python
student ={'name':'John','age':25,'courses':['Math','science']}

student['phone']='8010750902'
print(student.get('phone','Not found'))
```

```python
student ={'name':'John','age':25,'courses':['Math','science']}

student['name']='Aryan Gupta'
student['phone']='8010750902'
print(student)
```

```python
#update method
#we use this vwey we havce to upload the multiple values art a time
student ={'name':'John','age':25,'courses':['Math','science']}

student.update({'name':'Qt','age':89,'phone':8010509500})

print(student)
```
