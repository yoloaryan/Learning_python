# **List**

```python
courses =['History','Maths','Physics']
print(courses)
```

```python
courses =['History','Maths','Physics']
print(len(courses))
```

```python
courses =['History','Maths','Physics']
print(courses[1])
```

```python
courses =['History','Maths','Physics','Python']
print(courses[-1])
```

```python
courses =['History','Maths','Physics','Python']
print(courses[0:2])
```

```python
#append
courses =['History','Maths','Physics','Python']
courses.append('Art')
print(courses)
```

```python
#insert
courses =['History','Math','Physics','CompSci']

courses.insert(0,'Art')
print(courses)
```

```python
#extend
courses =['History','Math','Physics','Computer Science']
courses_2 =['Art','Education']

courses.extend(courses_2)
print(courses)
```

```python
#remove

courses =['History','Maths','Physics','Python']
courses.remove('Maths')
print(courses)
```

```python
#pop

courses =['History','Maths','Physics','Python']
courses.pop()
print(courses)
```

```python
#Sorting if the list
# reverse method
courses =['History','Maths','Physics','Python']
courses.reverse()
print(courses)
```

```python
#Sorting if the list

courses =['History','Maths','Physics','Python']
courses.sort()
print(courses)
```

```python
#Sorting if the list

courses =['History','Maths','Physics','Python']

nums =[1,2,4,8,9]

nums.sort(reverse=True)
courses.sort(reverse=True)
print(courses)
print(nums)
```

```python
#this method is used so. that we can not alter the original list

courses =['History','Maths','Physics','Python']

sorted_courses = sorted(courses)

print(sorted_courses)
```

```python
courses =['History','Maths','Physics','Python']

nums =[1,2,4,8,9]

print(min(nums))
```

```python
courses =['History','Maths','Physics','Python']

nums =[1,2,4,8,9]

print(max(nums))
```

```python
courses =['History','Math','Physics','Chemistry']
print(courses.index('Chemistry'))
```

```python
courses =['History','Math','Physics','Chemistry']
print('Art' in courses)
```

```python
#looping
courses =['History','Math','Physics','computer SCience']

for i in courses:
  print(i)
```

```python
#enumerate
courses =['History','Math','Physics','computer SCience']

for index,courses in enumerate(courses):
  print(index,courses)
```

```python

courses =['History','Math','Physics','computer SCience']

course_str = ' - '.join(courses)
new_list = course_str.split('_ - -')
print(course_str)
```
