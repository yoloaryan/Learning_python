```python
nums =[1,2,3,4,5]

for num in nums:
  if num == 3:
    print('Found!')
    break
  print(num)
```

```python
nums =[1,2,3,4,5]

for num in nums:
  if num == 3:
    print('Found!')
    continue
  print(num)
```

```python
#loop within the loop
nums =[1,2,3,4,5]

for num in nums:
  for letter in 'abc':
    print(num,letter)
```
